```python
#  add Q-learning agent
class QLearningAgent:
    def __init__(self, state_size, action_size, learning_rate=0.8, discount_factor=0.95, epsilon=0.2, epsilon_decay=0.995, epsilon_min=0.01):
        # توابع مربوط به agent Q-learning رو تعریف میکنه، مثل اندازه فضای حالت، تعداد اکشن ها، نرخ یادگیری، ضریب تنزیل و استراتژی اکتشاف
        self.state_size = state_size
        self.action_size = action_size
        self.learning_rate = learning_rate
        self.discount_factor = discount_factor
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min
        self.q_table = np.zeros((self.state_size, self.action_size)) # جدول Q-table رو با مقادیر اولیه صفر ایجاد میکنه

    def get_action(self, state):
        # تابعی که بر اساس استراتژی اکتشاف-بهره برداری، اکشن مناسب رو انتخاب میکنه
        if np.random.rand() <= self.epsilon:
            return np.random.choice(self.action_size) # اگر احتمال اکتشاف بالا باشه، یک اکشن تصادفی انتخاب میکنه
        else:
            return np.argmax(self.q_table[state]) # در غیر این صورت، اکشنی رو انتخاب میکنه که بیشترین مقدار Q-value رو دارد

    def update_q_table(self, state, action, reward, next_state):
        # تابعی که جدول Q-table رو بر اساس تجربه به روز میکنه
        future_max_q = np.max(self.q_table[next_state]) # بیشترین مقدار Q-value برای حالت بعدی رو محاسبه میکنه
        self.q_table[state][action] += self.learning_rate * (reward + self.discount_factor * future_max_q - self.q_table[state][action]) # فرمول به روزرسانی Q-value رو اعمال میکنه
        self.epsilon = max(self.epsilon_min, self.epsilon * self.epsilon_decay) # احتمال اکتشاف رو به تدریج کاهش میدهد
```

```python
# این بخش agent Q-learning رو آموزش میدهد
agent = QLearningAgent(state_size=X_train.shape[1], action_size=y_train["BORE_OIL_VOL"].unique().shape[0]) # یک agent Q-learning ایجاد میکنه و پارامترهای اولیه رو تنظیم میکنه
num_episodes = 1000 # تعداد دوره های آموزش رو مشخص میکنه

for episode in range(num_episodes):
    state = tuple(X_train.iloc[0]) # حالت اولیه رو به عنوان حالت شروع در نظر میگیرد
    done = False

    while not done:
        action = agent.get_action(state) # با استفاده از agent، یک اکشن مناسب رو انتخاب میکند
        next_state = tuple(X_train.iloc[action]) # حالت بعدی رو بر اساس اکشن انتخاب شده تعیین میکند
        reward = y_train.iloc[action] # پاداش (میزان تولید نفت) رو دریافت میکند
        agent.update_q_table(state, action, reward, next_state) # جدول Q-table رو به روز میکند
        state = next_state # حالت فعلی رو به عنوان حالت بعدی در نظر میگیرد
```
